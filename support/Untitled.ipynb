```python
def try_to_int(i):
    result = i
    try:
        result = int(i)
    except:
        pass
    finally:
        return str(result)
```

```python
df2 = pd.read_excel('API_NY.ADJ.NNTY.PC.CD_DS2_en_excel_v2_458.xls')
df2.columns = [try_to_int(i) for i in df2.iloc[2].tolist()]
df2 = df2.iloc[3:].reset_index(drop=True)
df2
```

```python
a = {i['Indicator Code']: i['Indicator Name'] for _, i in df1_cc.iterrows()}
b = {i['Indicator Code']: i['Indicator Name'] for _, i in df2_cc.iterrows()}
indicator_names = {**a, **b}
indicator_names
```

```python
def fill(df2, data=None):
    if data is None:
        data = {}
    years = df2.columns[df2.columns.str.isdigit()]
    for i, row in df2.iterrows():
        ccode = row['Country Code']
        cname = row['Country Name']
        icode = row['Indicator Code']
        iname = row['Indicator Name']    
        keyid = f"{ccode}_{icode}"
        for year in years:
            val = row[year]
            if pd.isna(val):
                continue
            year_key = int(year)
            data[(keyid, year_key)] = val ;"""{
                'value': val,
                'ccode': ccode,
                'cname': cname,
                'icode': icode,
                'iname': iname
            }"""
    return data
```

```python
data_df1 = fill(df1_cc)
data_df2 = fill(df2_cc)
```

```python
def data_to_df(data, fn=None):
    keys = sorted(data, key=lambda x: (x[1], x[0]))
    cols = sorted(set([i[0] for i in keys]))
    inds = sorted(set([int(i[1]) for i in keys]))
    df3 = pd.DataFrame(index=inds, columns=cols)
    for key in keys:
        df3.loc[key[1]][key[0]] = data[key]
    if fn:
        df3.to_csv(fn)
    return df3

def load_df(fn):
    return pd.read_csv(fn, index_col=0)

def store_df(data, fn):
    data_to_df(data, fn)
    return load_df(fn)

df3 = store_df(data_df1, 'df3.csv')
df3
```

```python
df4 = store_df(data_df2, 'df4.csv')
df4
```

```python
from os import makedirs
from matplotlib import pyplot as plt
```

```python
def to_png(df, col, path="plots2", size=(8, 4)):
    cmp1, cmp2 = col.split("_")
    nam1, nam2 = country_names[cmp1], indicator_names[cmp2]    
    ful1 = f"{cmp1} - {nam1}"
    ful2 = f"{cmp2} - {nam2}"
    ax = pd.DataFrame(df[col].tolist(), index=df.index, columns=[ful1]).plot.bar(title=ful2, figsize=size)
    fpath1 = f"{path}/by_country/{ful1}"
    fpath2 = f"{path}/by_param/{ful2}"    
    fpath3 = f"{path}/{nam1}"
    #makedirs(f"{path}/all", exist_ok=True)
    #makedirs(fpath1, exist_ok=True)
    #makedirs(fpath2, exist_ok=True)
    makedirs(fpath3, exist_ok=True)
    #ax.figure.savefig(f"{fpath1}/{col}.png")
    #ax.figure.savefig(f"{fpath2}/{col}.png")
    #ax.figure.savefig(f"{path}/all/{col}.png")
    ax.figure.savefig(f"{fpath3}/{col}.png")
    plt.close()
```
